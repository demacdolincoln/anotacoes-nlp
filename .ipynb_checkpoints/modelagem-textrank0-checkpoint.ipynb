```python
%pylab inline
import pickle
import gensim
import nltk
import networkx as nx
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm_notebook as tqdm
```

```python
figsize(8, 8)
```

```python
id2word = pickle.load(
    open("../../../datasets/skip_id2word.pickle", "rb")
)

word2id = pickle.load(
    open("../../../datasets/skip_word2id.pickle", "rb")
)

data = np.memmap("../../../datasets/skip_np.npy", mode="r+",
                    dtype=np.float, shape=(929605,100))
```

```python
text = pickle.load(
    open("corpus_text.pickle", "rb")
)

len(text), "".join([i+" " for i in text[2]])
```

```python
vocabulary = set()
for paragraph in text:
    for word in paragraph:
        if word in id2word:
            vocabulary.add(word2id[word])

vocabulary = np.array(list(vocabulary))
```

```python
similarity_matrix = np.zeros(shape=(vocabulary.size,
                             vocabulary.size), dtype=np.float)
```

```python
for i, x in tqdm(enumerate(vocabulary)):
    for j, y in enumerate(vocabulary):
        if i != j :
            similarity_matrix[i, j] = cosine_similarity(data[x].reshape(1, -1),
                                                        data[y].reshape(1, -1)).item(0)
```

```python
matshow(similarity_matrix)
```

```python
G = nx.from_numpy_array(similarity_matrix)
```

```python
scores = nx.pagerank(G)
```

```python
word_rank = sorted([(scores[i], s) for i, s in enumerate(vocabulary)], key=lambda x:x[0], reverse=True)
```

```python
for i in range(20):
    print(id2word[word_rank[i][1]])
```
