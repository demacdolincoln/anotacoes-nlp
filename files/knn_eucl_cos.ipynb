```python
data = np.random.randn(50, 2)
```

```python
scatter(data[:, 0], data[:, 1])
```

```python
pos = 26

p_cos, res_cos = knn(data, pos=pos)
p_cos_coord = data[p_cos]

p_eucl, res_eucl = knn(data, pos=pos, func="eucl")
p_eucl_coord = data[p_eucl]

fig, ax = subplots()


ax.scatter(data[:, 0], data[:, 1], c="gray", s=20)
ax.scatter(p_cos_coord[:, 0], p_cos_coord[:, 1], c="blue", s=200)
ax.scatter(p_eucl_coord[:, 0], p_eucl_coord[:, 1], c="red", s=200, alpha=0.5)
ax.scatter(data[pos, 0], data[pos, 1], c="yellow", s=130)

ax.grid(True, alpha=0.2)
ax.arrow(0, 0, data[pos, 0]*2, data[pos, 1]*2, color="yellow", overhang=1, head_width=0.05)
ax.set_xlim(0, 1.5)
ax.set_ylim(0, 1)

savefig("../images/eucl_vs_cos.png", dpi="figure", transparent=True)
```
